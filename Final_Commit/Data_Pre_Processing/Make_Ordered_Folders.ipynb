```python
import cv2, os, shutil
import numpy as np
from PIL import Image

# opening folders with data in them
flooding_true = "/content/drive/MyDrive/Third_Year_Project/Siamese_Data/output/train/FloodingTrue"
flooding_false = "/content/drive/MyDrive/Third_Year_Project/Siamese_Data/output/train/FloodingFalse"

paired = []
file_pairs = {}
for filename in os.listdir(flooding_true):

    temp = filename[0:4]
    if temp not in paired:

        flag = 0
        for filename2 in os.listdir(flooding_true):

            if filename2 != filename and filename2[0:4] == temp and flag == 0:
                flag = 1
                paired.append(temp)
                file_pairs[filename] = [filename2]
                break
        
        flag2 = 0
        for filename2 in os.listdir(flooding_false):

            if filename2[0:4] == temp and flag2 == 0 and flag == 1:
                flag2 = 1
                paired.append(temp)
                temp = file_pairs[filename]
                temp.append(filename2)
                file_pairs[filename] = temp
                break
        if filename in file_pairs:
          if len(file_pairs[filename]) < 2:
            file_pairs.pop(filename)

print('------------------------')
print(file_pairs)
print('------------------------')

count = 0
for i in file_pairs:

    #os.mkdir('./Image_Pairs_Equal/Pair' + str(count))
    if i in file_pairs:
      print(file_pairs[i][0])
      print(file_pairs[i][1])
      shutil.copy(flooding_true + '/' + i,'/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/PreProcessing/ANCHOR/img' + str(count) + '.png')
      shutil.copy(flooding_true + '/' + file_pairs[i][0],'/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/PreProcessing/POSITIVE/img' + str(count) + '.png')
      shutil.copy(flooding_false + '/' + file_pairs[i][1],'/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/PreProcessing/NEGATIVE/img' + str(count) + '.png')

      count += 1
```
