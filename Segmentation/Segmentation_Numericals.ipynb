```python
def get_data():

    gt = []
    rgb = []
    for e in sorted(os.listdir('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Temp_Images/RGB')):
        rgb.append(Image.open('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Temp_Images/RGB/' + e).resize((256, 256), Image.NEAREST).convert('RGB'))
        print(e)

    print('------------------------------------------')
    for e in sorted(os.listdir('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Temp_Images/GT')):
        gt.append(Image.open('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Temp_Images/GT/' + e).resize((256, 256), Image.NEAREST).convert('RGB'))
        print(e)
    return gt, rgb

input1, input2 = get_data()
print(input1)


def felzenszwalb_seg(images, images2):

    start = time.time()

    float_images = []
    segments_fz = []
    for e in images:

        #temp = img_as_float(e)
        temp = np.array(e)  

        float_images.append(temp)

        #segments_fz.append(felzenszwalb(temp, scale=400, sigma=0.75, min_size=0))
        segments_fz.append(felzenszwalb(temp, scale=450, sigma=0.5, min_size=75))

    float_images_gt = []
    for e in images2:
        temp = img_as_float(e)

        float_images_gt.append(temp)

    for num in segments_fz:
        print(f'Felzenszwalb number of segments: {len(np.unique(num))}')

    fig, ax = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax[0, 0].imshow(mark_boundaries(float_images[0], segments_fz[0]))
    ax[0, 0].set_title("IMG 1 - Felzenszwalbs RGB")
    ax[0, 1].imshow(mark_boundaries(float_images[1], segments_fz[1]))
    ax[0, 1].set_title('IMG 2 - Felzenszwalbs RGB')
    ax[0, 2].imshow(mark_boundaries(float_images[2], segments_fz[2]))
    ax[0, 2].set_title('IMG 3 - Felzenszwalbs RGB')
    ax[0, 3].imshow(mark_boundaries(float_images[3], segments_fz[3]))
    ax[0, 3].set_title('IMG 4 - Felzenszwalbs RGB')
    ax[1, 0].imshow(mark_boundaries(float_images[4], segments_fz[4]))
    ax[1, 0].set_title('IMG 5 - Felzenszwalbs RGB')
    ax[1, 1].imshow(mark_boundaries(float_images[5], segments_fz[5]))
    ax[1, 1].set_title('IMG 6 - Felzenszwalbs RGB')
    ax[1, 2].imshow(mark_boundaries(float_images[6], segments_fz[6]))
    ax[1, 2].set_title('IMG 7 - Felzenszwalbs RGB')
    ax[1, 3].imshow(mark_boundaries(float_images[7], segments_fz[7]))
    ax[1, 3].set_title('IMG 8 - Felzenszwalbs RGB')

    plt.savefig("RGB_FZ.png")

    fig2, ax2 = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax2[0, 0].imshow(mark_boundaries(float_images_gt[0], segments_fz[0], color = (0,0,0)))
    ax2[0, 0].set_title("IMG 1 - Felzenszwalbs GT")
    ax2[0, 1].imshow(mark_boundaries(float_images_gt[1], segments_fz[1], color = (0,0,0)))
    ax2[0, 1].set_title("IMG 2 - Felzenszwalbs GT")
    ax2[0, 2].imshow(mark_boundaries(float_images_gt[2], segments_fz[2], color = (0,0,0)))
    ax2[0, 2].set_title("IMG 3 - Felzenszwalbs GT")
    ax2[0, 3].imshow(mark_boundaries(float_images_gt[3], segments_fz[3], color = (0,0,0)))
    ax2[0, 3].set_title("IMG 4 - Felzenszwalbs GT")
    ax2[1, 0].imshow(mark_boundaries(float_images_gt[4], segments_fz[4], color = (0,0,0)))
    ax2[1, 0].set_title("IMG 5 - Felzenszwalbs GT")
    ax2[1, 1].imshow(mark_boundaries(float_images_gt[5], segments_fz[5], color = (0,0,0)))
    ax2[1, 1].set_title("IMG 6 - Felzenszwalbs GT")
    ax2[1, 2].imshow(mark_boundaries(float_images_gt[6], segments_fz[6], color = (0,0,0)))
    ax2[1, 2].set_title("IMG 7 - Felzenszwalbs GT")
    ax2[1, 3].imshow(mark_boundaries(float_images_gt[7], segments_fz[7], color = (0,0,0)))
    ax2[1, 3].set_title("IMG 8 - Felzenszwalbs GT")

    plt.savefig("GT_FZ.png")

    for a in ax.ravel():
        a.set_axis_off()

    for b in ax2.ravel():
        b.set_axis_off()

    plt.tight_layout()
    plt.show()

    end = time.time()
    print('Felzenszwalbs elapsed time:')
    print(end - start)

def slic_seg(images, images2):

    start = time.time()

    float_images = []
    segments_fz = []
    for e in images:

        #temp = img_as_float(e)
        temp = np.array(e)

        float_images.append(temp)

        segments_fz.append(slic(temp, n_segments=250, compactness=10, sigma=1,
                         start_label=1))

    float_images_gt = []
    for e in images2:
        temp = img_as_float(e)

        float_images_gt.append(temp)

    for num in segments_fz:
        print(f'SLIC number of segments: {len(np.unique(num))}')

    fig, ax = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax[0, 0].imshow(mark_boundaries(float_images[0], segments_fz[0]))
    ax[0, 0].set_title("IMG 1 - SLIC RGB")
    ax[0, 1].imshow(mark_boundaries(float_images[1], segments_fz[1]))
    ax[0, 1].set_title('IMG 2 - SLIC RGB')
    ax[0, 2].imshow(mark_boundaries(float_images[2], segments_fz[2]))
    ax[0, 2].set_title('IMG 3 - SLIC RGB')
    ax[0, 3].imshow(mark_boundaries(float_images[3], segments_fz[3]))
    ax[0, 3].set_title('IMG 4 - SLIC RGB')
    ax[1, 0].imshow(mark_boundaries(float_images[4], segments_fz[4]))
    ax[1, 0].set_title('IMG 5 - SLIC RGB')
    ax[1, 1].imshow(mark_boundaries(float_images[5], segments_fz[5]))
    ax[1, 1].set_title('IMG 6 - SLIC RGB')
    ax[1, 2].imshow(mark_boundaries(float_images[6], segments_fz[6]))
    ax[1, 2].set_title('IMG 7 - SLIC RGB')
    ax[1, 3].imshow(mark_boundaries(float_images[7], segments_fz[7]))
    ax[1, 3].set_title('IMG 8 - SLIC RGB')

    plt.savefig("RGB_SLIC.png")

    fig2, ax2 = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax2[0, 0].imshow(mark_boundaries(float_images_gt[0], segments_fz[0], color = (0,0,0)))
    ax2[0, 0].set_title("IMG 1 - SLIC GT")
    ax2[0, 1].imshow(mark_boundaries(float_images_gt[1], segments_fz[1], color = (0,0,0)))
    ax2[0, 1].set_title("IMG 2 - SLIC GT")
    ax2[0, 2].imshow(mark_boundaries(float_images_gt[2], segments_fz[2], color = (0,0,0)))
    ax2[0, 2].set_title("IMG 3 - SLIC GT")
    ax2[0, 3].imshow(mark_boundaries(float_images_gt[3], segments_fz[3], color = (0,0,0)))
    ax2[0, 3].set_title("IMG 4 - SLIC GT")
    ax2[1, 0].imshow(mark_boundaries(float_images_gt[4], segments_fz[4], color = (0,0,0)))
    ax2[1, 0].set_title("IMG 5 - SLIC GT")
    ax2[1, 1].imshow(mark_boundaries(float_images_gt[5], segments_fz[5], color = (0,0,0)))
    ax2[1, 1].set_title("IMG 6 - SLIC GT")
    ax2[1, 2].imshow(mark_boundaries(float_images_gt[6], segments_fz[6], color = (0,0,0)))
    ax2[1, 2].set_title("IMG 7 - SLIC GT")
    ax2[1, 3].imshow(mark_boundaries(float_images_gt[7], segments_fz[7], color = (0,0,0)))
    ax2[1, 3].set_title("IMG 8 - SLIC GT")

    plt.savefig("GT_SLIC.png")

    for a in ax.ravel():
        a.set_axis_off()

    for b in ax2.ravel():
        b.set_axis_off()

    plt.tight_layout()
    plt.tight_layout()
    plt.show()

    end = time.time()
    print('SLIC elapsed time:')
    print(end - start)

def quickshift_seg(images, images2):

    start = time.time()

    float_images = []
    segments_fz = []
    for e in images:

        #temp = img_as_float(e)
        temp = np.array(e)

        float_images.append(temp)

        #segments_fz.append(quickshift(temp,kernel_size=3, max_dist=50, ratio=0.75))
        segments_fz.append(quickshift(temp, kernel_size=5, max_dist=15, ratio=0.90))

    float_images_gt = []
    for e in images2:
        temp = img_as_float(e)

        float_images_gt.append(temp)

    for num in segments_fz:
        print(f'Quickshift number of segments: {len(np.unique(num))}')

    fig, ax = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax[0, 0].imshow(mark_boundaries(float_images[0], segments_fz[0]))
    ax[0, 0].set_title("IMG 1 - Quickshift RGB")
    ax[0, 1].imshow(mark_boundaries(float_images[1], segments_fz[1]))
    ax[0, 1].set_title('IMG 2 - Quickshift RGB')
    ax[0, 2].imshow(mark_boundaries(float_images[2], segments_fz[2]))
    ax[0, 2].set_title('IMG 3 - Quickshift RGB')
    ax[0, 3].imshow(mark_boundaries(float_images[3], segments_fz[3]))
    ax[0, 3].set_title('IMG 4 - Quickshift RGB')
    ax[1, 0].imshow(mark_boundaries(float_images[4], segments_fz[4]))
    ax[1, 0].set_title('IMG 5 - Quickshift RGB')
    ax[1, 1].imshow(mark_boundaries(float_images[5], segments_fz[5]))
    ax[1, 1].set_title('IMG 6 - Quickshift RGB')
    ax[1, 2].imshow(mark_boundaries(float_images[6], segments_fz[6]))
    ax[1, 2].set_title('IMG 7 - Quickshift RGB')
    ax[1, 3].imshow(mark_boundaries(float_images[7], segments_fz[7]))
    ax[1, 3].set_title('IMG 8 - Quickshift RGB')

    plt.savefig("RGB_QS.png")

    fig2, ax2 = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax2[0, 0].imshow(mark_boundaries(float_images_gt[0], segments_fz[0], color = (0,0,0)))
    ax2[0, 0].set_title("IMG 1 - Quickshift GT")
    ax2[0, 1].imshow(mark_boundaries(float_images_gt[1], segments_fz[1], color = (0,0,0)))
    ax2[0, 1].set_title("IMG 2 - Quickshift GT")
    ax2[0, 2].imshow(mark_boundaries(float_images_gt[2], segments_fz[2], color = (0,0,0)))
    ax2[0, 2].set_title("IMG 3 - Quickshift GT")
    ax2[0, 3].imshow(mark_boundaries(float_images_gt[3], segments_fz[3], color = (0,0,0)))
    ax2[0, 3].set_title("IMG 4 - Quickshift GT")
    ax2[1, 0].imshow(mark_boundaries(float_images_gt[4], segments_fz[4], color = (0,0,0)))
    ax2[1, 0].set_title("IMG 5 - Quickshift GT")
    ax2[1, 1].imshow(mark_boundaries(float_images_gt[5], segments_fz[5], color = (0,0,0)))
    ax2[1, 1].set_title("IMG 6 - Quickshift GT")
    ax2[1, 2].imshow(mark_boundaries(float_images_gt[6], segments_fz[6], color = (0,0,0)))
    ax2[1, 2].set_title("IMG 7 - Quickshift GT")
    ax2[1, 3].imshow(mark_boundaries(float_images_gt[7], segments_fz[7], color = (0,0,0)))
    ax2[1, 3].set_title("IMG 8 - Quickshift GT")

    plt.savefig("GT_QS.png")

    for a in ax.ravel():
        a.set_axis_off()

    for b in ax2.ravel():
        b.set_axis_off()

    plt.tight_layout()
    plt.tight_layout()
    plt.show()

    end = time.time()
    print('Quickshift elapsed time:')
    print(end - start)

def watershed_seg(images, images2):

    import numpy as np
    from skimage import io, segmentation
    import matplotlib.pyplot as plt

    start = time.time()

    float_images = []
    segments_fz = []
    for e in images:

        #temp = img_as_float(e)
        temp = np.array(e)

        float_images.append(temp)

        gradient = sobel(rgb2gray(temp))
        segments_fz.append(watershed(gradient, markers=250, compactness=0.001))

    float_images_gt = []
    for e in images2:
        temp = img_as_float(e)

        float_images_gt.append(temp)

    for num in segments_fz:
        print(f'Watershed number of segments: {len(np.unique(num))}')

    fig, ax = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax[0, 0].imshow(mark_boundaries(float_images[0], segments_fz[0]))
    ax[0, 0].set_title("IMG 1 - Compact Watershed RGB")
    ax[0, 1].imshow(mark_boundaries(float_images[1], segments_fz[1]))
    ax[0, 1].set_title('IMG 2 - Compact Watershed RGB')
    ax[0, 2].imshow(mark_boundaries(float_images[2], segments_fz[2]))
    ax[0, 2].set_title('IMG 3 - Compact Watershed RGB')
    ax[0, 3].imshow(mark_boundaries(float_images[3], segments_fz[3]))
    ax[0, 3].set_title('IMG 4 - Compact Watershed RGB')
    ax[1, 0].imshow(mark_boundaries(float_images[4], segments_fz[4]))
    ax[1, 0].set_title('IMG 5 - Compact Watershed RGB')
    ax[1, 1].imshow(mark_boundaries(float_images[5], segments_fz[5]))
    ax[1, 1].set_title('IMG 6 - Compact Watershed RGB')
    ax[1, 2].imshow(mark_boundaries(float_images[6], segments_fz[6]))
    ax[1, 2].set_title('IMG 7 - Compact Watershed RGB')
    ax[1, 3].imshow(mark_boundaries(float_images[7], segments_fz[7]))
    ax[1, 3].set_title('IMG 8 - Compact Watershed RGB')

    plt.savefig("RGB_CW.png")

    fig2, ax2 = plt.subplots(2, 4, figsize=(20, 20), sharex=True, sharey=True)

    ax2[0, 0].imshow(mark_boundaries(float_images_gt[0], segments_fz[0], color = (0,0,0)))
    ax2[0, 0].set_title("IMG 1 - Compact Watershed GT")
    ax2[0, 1].imshow(mark_boundaries(float_images_gt[1], segments_fz[1], color = (0,0,0)))
    ax2[0, 1].set_title("IMG 2 - Compact Watershed GT")
    ax2[0, 2].imshow(mark_boundaries(float_images_gt[2], segments_fz[2], color = (0,0,0)))
    ax2[0, 2].set_title("IMG 3 - Compact Watershed GT")
    ax2[0, 3].imshow(mark_boundaries(float_images_gt[3], segments_fz[3], color = (0,0,0)))
    ax2[0, 3].set_title("IMG 4 - Compact Watershed GT")
    ax2[1, 0].imshow(mark_boundaries(float_images_gt[4], segments_fz[4], color = (0,0,0)))
    ax2[1, 0].set_title("IMG 5 - Compact Watershed GT")
    ax2[1, 1].imshow(mark_boundaries(float_images_gt[5], segments_fz[5], color = (0,0,0)))
    ax2[1, 1].set_title("IMG 6 - Compact Watershed GT")
    ax2[1, 2].imshow(mark_boundaries(float_images_gt[6], segments_fz[6], color = (0,0,0)))
    ax2[1, 2].set_title("IMG 7 - Compact Watershed GT")
    ax2[1, 3].imshow(mark_boundaries(float_images_gt[7], segments_fz[7], color = (0,0,0)))
    ax2[1, 3].set_title("IMG 8 - Compact Watershed GT")

    plt.savefig("GT_CW.png")

    for a in ax.ravel():
        a.set_axis_off()

    for b in ax2.ravel():
        b.set_axis_off()

    plt.tight_layout()
    plt.tight_layout()
    plt.show()

    end = time.time()
    print('Watershed elapsed time:')
    print(end - start)


watershed_seg(input2, input1)
slic_seg(input2, input1)
felzenszwalb_seg(input2, input1)
quickshift_seg(input2, input1)
```

```python
from skimage.util.dtype import img_as_int, img_as_ubyte

'''
(253, 231, 36) -> Cloud GT color code
(53, 183, 120) -> Land GT color code
(48, 103, 141) -> Water GT color code
'''
def test(method, path, path2, iter_num):

    import numpy as np
    from skimage import io
    from skimage.transform import rescale, resize
    import matplotlib.pyplot as plt

    #img = io.imread(path)
    #img2 = io.imread(path2)

    img = Image.open(path).resize((256, 256), Image.NEAREST).convert('RGB')#.resize((256, 256), Image.ANTIALIAS)
    img = np.array(img)

    img2 = Image.open(path2).resize((256, 256), Image.NEAREST).convert('RGB')#.resize((256, 256), Image.ANTIALIAS)
    img2 = np.array(img2)

    #img = img_as_float(path)
    #img2 = img_as_float(path2)
    #img = img_as_ubyte(path)
    #img2 = img_as_ubyte(path2)

    #print(img2)

    if method == 'Felzenszwalbs':
      #segments = segmentation.felzenszwalb(img, scale=450, sigma=0.50, min_size=0)
      segments = felzenszwalb(img, scale=450, sigma=0.5, min_size=75)
      print(f'Felzenszwalb number of segments: {len(np.unique(segments))}')

    elif method == 'SLIC':
      segments = slic(img, n_segments=250, compactness=10, sigma=1, start_label=1)
      print(f'SLIC number of segments: {len(np.unique(segments))}')

    elif method == 'CompactWatershed':
      gradient = sobel(rgb2gray(img))
      segments = watershed(gradient, markers=250, compactness=0.001)
      print(f'Watershed number of segments: {len(np.unique(segments))}')

    elif method == 'Quickshift':
      #segments = segmentation.quickshift(img, kernel_size=10, max_dist=250, ratio=0.75)
      segments = quickshift(img, kernel_size=5, max_dist=15, ratio=0.90)
      print(f'Quickshift number of segments: {len(np.unique(segments))}')

    else:
      segments = slic(img, n_segments=250, compactness=10, sigma=1, start_label=1)

    total_superpixels = len(np.unique(segments))
    pure_land = 0
    pure_cloud = 0
    pure_water = 0

    #print(get_colors_in_img(path2))
    #print(img2)

    for index in np.unique(segments):
        
        segment = img.copy()
        segment[segments != index] = 0

        segment2 = img2.copy()
        segment2[segments != index] = 0

        plt.imsave('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Segments_All/Segments_' + str(method) + '/RGB/img' + str(iter_num) + '_segment_' + str(index) + '.png', segment)
        plt.imsave('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Segments_All/Segments_' + str(method) + '/GT/img' + str(iter_num) +'_segment_' + str(index) + '.png', segment2)
        
        land_purity, cloud_purity, water_purity =  count_land_cloud_and_water('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Segments_All/Segments_' + str(method) + '/GT/img' + str(iter_num) +'_segment_' + str(index) + '.png')
        if land_purity >= 0.9:
          pure_land += 1

        elif cloud_purity >= 0.9:
          pure_cloud += 1

        elif water_purity >= 0.9:
          pure_water += 1

    print('Land pure pixels are: ' + str(pure_land))
    print('Water pure pixels are: ' + str(pure_water))
    print('Cloud pure pixels are: ' + str(pure_cloud))

    with open('/content/drive/MyDrive/Third_Year_Project/PyCharm Folder/Sk-img segmentation/Segments_All/Segments_' + str(method) + '/segments_data.csv', 'a', newline ='') as csvfile:
      
      header = ['Image', 'Water Pure SPs', 'Land Pure SPs', 'Cloud Pure SPs', 'Total SP']
      data = [str(path), pure_water, pure_land, pure_cloud, total_superpixels]

      writer = csv.writer(csvfile)

      # write the header
      writer.writerow(header)

      # write the data
      writer.writerow(data)

      csvfile.close()

def get_colors_in_img(path):
    img = Image.open(path)
    colors = img.convert('RGB').getcolors()

    print(colors)

from skimage import io, segmentation
from skimage.transform import rescale, resize
def count_land_cloud_and_water(path):

    #im = Image.open(path)
    im = Image.open(path).resize((256, 256), Image.NEAREST).convert('RGB')

    #im = resize(io.imread(path), (256,256))

    #print(im.getdata())

    land = 0
    water = 0
    cloud = 0
    black = 0
    total = 0

    for pixel in im.getdata():  
        #print(pixel)
        total += 1
        if pixel == (53, 183, 120):  # if your image is RGB (if RGBA, (0, 0, 0, 255) or so
            cloud += 1
        elif pixel == (253, 231, 36):
            land += 1
        elif pixel == (48, 103, 141):
            water += 1
        elif pixel == (0, 0, 0):
            black += 1

    land_purity = land/(total - black)
    pure_land = 0

    cloud_purity = cloud/(total - black)
    pure_cloud = 0

    water_purity = water/(total - black)
    pure_water = 0

    #print('land=' + str(land) + ', cloud=' + str(cloud) + ', water=' + str(water) + ', black=' + str(black))
    #print('land %=' + str(land/(total - black)) + ', cloud %=' + str(cloud/(total - black)) + ', water %=' + str(water/(total - black)))
    #print("----------------------------------------------------")

    return land_purity, cloud_purity, water_purity
```
